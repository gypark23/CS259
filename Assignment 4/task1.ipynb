First, read in the data

```python
import pandas as pd
import csv

data = pd.read_csv("data.csv")
data_full = pd.read_csv("data-full.csv")
```

Iterate through data and data_full to only include rows with empty title.
Make a separate dataframe for comparison.

```python
data_title = pd.read_csv("data.csv")
data_full_title = pd.read_csv("data-full.csv")
for i in data.index:
    if(pd.isnull(data.iloc[i,1]) == False):
        data_title = data_title.drop(index = i)
        data_full_title = data_full_title.drop(index = i)
```

There are total 57 NaN in titles, and we look at them and compare to identify patterns.

```python
data_title
```

```python
data_full_title
```

```python
data_title['author'].value_counts()
```

We can see a lot of Rob Micthum's song title are removed, constituting over half of the total title removed songs. Because the probability of missing title is related to the author, this would be MAR, missing at random, since probability of missing data in title depends on author, but not the values of title.

We do the same for score:

```python
data_score = pd.read_csv("data.csv")
data_full_score = pd.read_csv("data-full.csv")
for i in data.index:
    if(pd.isnull(data.iloc[i,4]) == False):
        data_score = data_score.drop(index = i)
        data_full_score = data_full_score.drop(index = i)
```

```python
data_score
```

```python
data_full_score
```

```python
data_full_score['score'].value_counts()
```

We can see that score of 6.6 are all completely removed. This would be MNAR, Missing Not At Random, since probability of missing data score is related to the values of score.

Now, we only look at data_score and data_title.

```python
data_title
```

It is easy to conclude that the missing data is still depending on the author name, with the same characterization of MAR.

```python
data_score
data_score['artist'].value_counts()
data_score['author'].value_counts()
```

However, we can no longer identify the pattern in data_score, therefore it would be MCAR, Missing Completely at Random without the original data.
