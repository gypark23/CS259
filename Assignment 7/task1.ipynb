```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

Randomized Response for 1000 people with 50%, 25%, 10%, 1% probability of saying yes (1)

```python
for probability in [0.5, 0.25, 0.1, 0.01]:    
    proportions = []
    for i in range(100):    
        arr = np.random.choice(2, 1000, p = [1 - probability, probability])
        responsearr = []
        for response in arr:
            coinflip = random.randint(0, 1)
            #heads 1 tails 0
            if(coinflip):
                responsearr.append(response)
            else:
                coinflip = random.randint(0, 1)
                #heads 1 tails 0
                responsearr.append(coinflip)
        proportions.append(responsearr.count(1) / len(responsearr))

    plt.hist(proportions)
    plt.title("Randomized Response Histogram of Probability of " + str(probability * 100) + "%")
    plt.savefig("plots/" + str(probability) + ".png", dpi = 300)
    plt.show()
```

Randomized response but with different coin flip probability of 1/8, 1/4 and 3/4

```python
for coinprobability in [1/8, 1/4, 3/4]:
    for probability in [0.5, 0.25, 0.1, 0.01]:    
        proportions = []
        for i in range(100):    
            arr = np.random.choice(2, 1000, p = [1 - probability, probability])
            responsearr = []
            for response in arr:
                coinflip = 0 if random.random() < coinprobability else 1
                #heads 1 tails 0
                if(coinflip):
                    responsearr.append(response)
                else:
                    coinflip = random.randint(0, 1)
                    #heads 1 tails 0
                    responsearr.append(coinflip)
            proportions.append(responsearr.count(1) / len(responsearr))

        plt.hist(proportions)
        plt.title("Randomized Response Histogram of Probability of " + str(probability * 100) + "% with Coin Flip Probability of " + str(coinprobability * 100) + "%", fontsize = 8)
        plt.savefig("plots/" + str(coinprobability) + str(probability) + ".png", dpi = 500)
        plt.show()
```
